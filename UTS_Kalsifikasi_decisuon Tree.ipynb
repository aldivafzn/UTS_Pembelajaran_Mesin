## Preprosesing Data

```python
from sklearn.preprocessing import LabelEncoder

# Assuming 'hasil' is your DataFrame
for col in ['age', 'job', 'marital', 'education', 'default', 'balance',
            'housing', 'loan', 'contact', 'day', 'month', 'duration',
            'campaign', 'pdays', 'previous', 'poutcome','y']:
    # Check if the column exists in the DataFrame
    if col in hasil.columns:
        # Check the data type before applying LabelEncoder
        if hasil[col].dtype == 'object':
            label_encoder = LabelEncoder()
            hasil[col] = label_encoder.fit_transform(hasil[col])
        else:
            # Handle other data types (if needed)
            print(f"Skipping encoding for non-object column: {col}")
    else:
        print(f"Column '{col}' not found in the DataFrame.")

#aldivafadliefauzan_202131017
```

## Memisahkan Kolom Target

## Splitting Data Testting dan Traning?

```python
from sklearn.model_selection import train_test_split

#aldivafadliefauzan_202131017
```

## 80 : 20

```python
x_train, x_test, y_train, y_test = train_test_split (x, y, test_size = 0.2)

#aldivafadliefauzan_202131017
```

#### Desision Tree

```python
from sklearn.tree import DecisionTreeClassifier
modelDT = DecisionTreeClassifier (max_depth = 5)
modelDT.fit (x_train, y_train)

#aldivafadliefauzan_202131017
```

```python
# Testing model untuk melakukan Prediksi
hasil_prediksi = modelDT.predict (x_test)

#aldivafadliefauzan_202131017
```

```python
from sklearn.metrics import accuracy_score
hasil_prediksi_dt = modelDT.predict (x_test)
print ('Akurasi Model : ', accuracy_score (y_test, hasil_prediksi_dt)*100)

#aldivafadliefauzan_202131017
```

## Test Data

```python
data_baru = ([[71,5,0,0,0,1729,0,0,0,17,9,456,2,-1,0,3]])
hasil_prediksi = modelDT.predict (data_baru)
if hasil_prediksi == 1:
    print("Pelanggan Tertarik")
elif hasil_prediksi == 0:
    print("Pelanggan Tidak Tetarik")
else:
    print("Hasil prediksi tidak valid")
    
    #aldivafadliefauzan_202131017
```
